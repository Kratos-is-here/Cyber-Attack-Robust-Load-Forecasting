```python
model = nn.Sequential(
    nn.Linear(2, 50),
    nn.Linear(50, 50),
    nn.Linear(50, 1)
)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
criterion = nn.L1Loss()
optim = optim.SGD(model.parameters(), lr = 0.01)
```

```python
class load_data(DataLoader):
      def __init__(self, csv_file, transform=None,attack='scaling'):
        """
        Args:
            csv_file (string): Path to the csv file with annotations.
            root_dir (string): Directory with all the images.
            transform (callable, optional): Optional transform to be applied
                on a sample.
        """
        self.dataset = pd.read_csv(csv_file)
        self.transform = transform
        self.attack=attack
      def __len__(self):
        return len(self.dataset)

      def __getitem__(self, idx):
        if self.attack=='scaling':
            c=5
        if self.attack=='ramping':
            c=7
        if self.attack=='random':
            c=9


        data= self.dataset.iloc[idx,c]
        label= self.dataset.iloc[idx,1+c]

        
        return (data,label)
```
